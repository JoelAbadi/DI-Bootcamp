```python
from openpyxl import load_workbook
file_name= 'Plants.xlsx'

wb = load_workbook(filename=file_name)

ws = wb.active

row = 2

while True:
  plant_name = ws.cell(row=row, column=1).value
  if plant_name is None:
    break

  in_stock= ws.cell(row=row, column=8).value

  if in_stock is not None and in_stock == 'No':
    print(f'not in stock: {plant_name}')

  row += 1
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("productSales.xlsx")

grouped_df = df.groupby('product')['sales'].sum().reset_index()

plt.figure(figsize=(8, 5))
plt.bar(grouped_df['product'], grouped_df['sales'], color='skyblue')
plt.title("Total Sales per Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
output_file = "sales_report.xlsx"

with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    grouped_df.to_excel(writer, sheet_name='Summary', index=False)

files.download(output_file)
```
